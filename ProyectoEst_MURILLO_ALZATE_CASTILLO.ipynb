```python
#Se realiza un gráfico de caja para realizar una presentación estadística clave como lo son la mediana, la media y los cuartiles y los aspectos de distribución de los datos.
#En este caso se evaluan los precios de los moviles con respecto a su cobertura 4G/LTE, los que si la tienen, (tienen una mediana, cuartil superior, cuartil inferior y media>25000) y algunos tienen valores externos superiores a 250000. 
#Los moviles que no tienen cobertura 4G/lte son menos por lo tanto el diagrama es más pequeño y se puede notar que son más económicos que los que si la tienen debido a los resultados estadisticos como la media, mediana, entre otros que se encuentran aún más abajo que el diagrama de si

sns.boxplot(x=cel['4G/ LTE'], y=cel['Price'],showmeans=True)
```

```python
#En este cuadro se agrupan todos los datos por lo tanto el resultado es 666 filas y 11 columnas, muchas menos que en los cuadros anteriores
cel.groupby(by=['Operating system','Price']).mean()
```

```python
#Por medio de este diagrama de caja podemos ver la relación de la capacidad de la bateria según el sistema operativo del movil, en este caso el valor de sus medianas es diferente ya que las capacidades de los sistemas operativos varian, pero sus diferencias no están muy lejanas.
sns.boxplot(x=cel['Operating system'], y=cel['Battery capacity (mAh)'],showmeans=True)
```

```python
#Se apilan los datos por medio de un gráfico de densidad de kernel
sns.kdeplot(x='Price', hue ='3G', data=cel, shade=True);
```

```python
#Se realizo el diagrama de linea para ver la pendiente y los intervalos de confianza que posee la relación precio con el tamaño de la pantalla de los celulares y se determino que el precio del movil es mucho mayor si su pantalla es más grande
sns.lmplot(x="Price", y="Screen size (inches)",data=cel);
```

```python
#Por medio de este diagrama de caja podemos observar la relación entre el precio y el sistema operativo, en el cual resaltaremos que el iOS es uno de los sistemas que cuenta con una gran diferencia entre sus precios y donde su mediana esta entre 25.000 y 50.000 
sns.boxplot(x=cel['Price'], y=cel['Operating system'],showmeans=True)
```

```python
#Por medio de este diagrama de dispersión con relación Precio y la capacidad de la bateria. Según este se puede inferir que se denota la existencia de una estrecha relación entre la capacidad de la bateria y su precio, por lo tanto en la mayor parte de los casos entre más mAh tenga el movil más caro será
ax = sns.scatterplot(x="Battery capacity (mAh)", y="Price", data=cel);
ax.set_title("Rear camera vs Price")
```

```python
#Por medio de este diagrama de barras se puede ver claramente la relación de l marca del movil con el precio, por lo tanto los celulares más caros son los Android con precios de hasta 175.000 y los iOS con precios mayores a 100.000
plt.bar("Operating system","Price", data = cel, color = "green")
plt.xlabel("sistema operativo")
plt.ylabel("precio")
plt.title("Sistema operativo vs precio")
plt.show()
```

```python
#Se realiza el diagrama de torta para poder identificar la cantidad de datos según las capacidades de bateria de los moviles, pero como son bastantes el diagrama no se puede observar muy bien

plt.pie( "Battery capacity (mAh)",data=cel)
plt.show()
```
